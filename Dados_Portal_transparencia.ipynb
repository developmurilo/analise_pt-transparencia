```python
# Converte colunas de string para float, ajustando vírgulas para pontos
df_viagens['Valor passagens'] = df_viagens['Valor passagens'].str.replace(",",".").astype(float)
df_viagens['Valor devolução'] = df_viagens['Valor devolução'].str.replace(",",".").astype(float)
df_viagens['Valor outros gastos'] = df_viagens['Valor outros gastos'].str.replace(",",".").astype(float)
```

```python
# Soma os valores de diárias e passagens
df_viagens['Valor diárias'] + df_viagens['Valor passagens']
```

```python
df_viagens['Despesas Total'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos'] #Criando uma nova coluna chamada "Despesas Total" onde eu somo todas as despesas Diarias, passagens e outros gastos
```

```python
# Calcula a proporção (%) de viagens por cargo
(df_viagens['Cargo'].value_counts(normalize=True) *100).rename("Proporção de Viajens").reset_index()
```

```python
# Define o formato de exibição de floats com 2 casas decimais
pd.set_option('display.float_format', '{:.2f}'.format)

# Calcula a média das despesas totais por cargo e ordena do maior para o menor
df_viagens.groupby("Cargo")["Despesas Total"].mean().reset_index().sort_values(by="Despesas Total", ascending=False)
```

```python
# Aplica o filtro e exibe apenas os cargos com mais de 1% das viagens
viagens_por_cargo[filtro_mais_de_1_porcento]
```

```python
# Aplica o filtro para exibir cargos que começam com 'TECNICO'
viagens_por_cargo[filtro_e_tecnico]
```

```python
# Aplica filtros para exibir cargos que começam com 'TECNICO' ou têm mais de 1% das viagens
viagens_por_cargo[filtro_e_tecnico | filtro_mais_de_1_porcento]
```

```python
# Converte a coluna 'Período - Data de início' para o formato datetime
pd.to_datetime(df_viagens["Período - Data de início"], format="%d/%m/%Y")
```

```python
# Converte as colunas de data para o formato datetime
df_viagens["Período - Data de início"] = pd.to_datetime(df_viagens["Período - Data de início"], format="%d/%m/%Y")
df_viagens["Período - Data de fim"] = pd.to_datetime(df_viagens["Período - Data de fim"], format="%d/%m/%Y")
```

```python
df_viagens
```

```python
# Cria uma nova coluna com o nome do mês da data de início da viagem
df_viagens["Mes da viajem"] = df_viagens['Período - Data de início'].dt.month_name()
```

```python
# Calcula a duração da viagem em dias e armazena na nova coluna 'Dias de viajem'
df_viagens["Dias de viajem"] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']).dt.days
```

```python
#Criando segregação com novas colunas
# Segrega os dados por cargo, calculando a média de despesas, média de duração de viagem,
# total de despesas por cargo, destinos mais frequentes e o número de viagens por cargo
 (

    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media=("Despesas Total", "mean" ),
        duração_media=("Dias de viajem", "mean"),
        despesas_totais_por_cargo=("Despesas Total", "sum"),
        destinos_mais_frequantes=("Destinos",pd.Series.mode),
        numero_viajens_por_cargo=("Cargo", "count")
        )
    .reset_index()
)
```

```python
# Calcula a proporção de ocorrências de cada cargo e armazena em um novo DataFrame
df_cargos = df_viagens['Cargo'].value_counts(normalize=True).reset_index()
df_cargos
```

```python
# Filtra cargos com proporção superior a 1% no DataFrame df_cargos
df_cargos.loc[df_cargos['proportion'] > 0.01, "Cargo"]
```

```python
# Consolida os dados por cargo, calculando média de despesas, média de duração, total de despesas,
# destinos mais frequentes e número de viagens por cargo
df_viajens_consolidados = (

    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media=("Despesas Total", "mean" ),
        duração_media=("Dias de viajem", "mean"),
        despesas_totais_por_cargo=("Despesas Total", "sum"),
        destinos_mais_frequantes=("Destinos",pd.Series.mode),
        numero_viajens_por_cargo=("Cargo", "count")
        )
    .reset_index()
)
```

```python
# Filtra os cargos com proporção superior a 1% e armazena os cargos mais relevantes
cargos_mais_relevantes = df_cargos.loc[df_cargos['proportion'] > 0.01, "Cargo"]
```

```python
# Cria um filtro para selecionar cargos presentes na lista de cargos mais relevantes
filtro = df_viajens_consolidados['Cargo'].isin(cargos_mais_relevantes)
```

```python
# Aplica o filtro para criar o DataFrame com os cargos mais relevantes
df_final = df_viajens_consolidados[filtro]
```

```python
df_final
```

```python
# Ordena os dados por número de viagens por cargo em ordem decrescente e plota um gráfico de barras
df_final = df_final.sort_values(by="numero_viajens_por_cargo", ascending=False)
df_final.plot(x="Cargo", y="numero_viajens_por_cargo", kind="bar")
```

## Criação de graficos para analise


```python
# Importa a biblioteca Matplotlib para criação de gráficos
import matplotlib.pyplot as plt
```

```python
# Cria um gráfico de barras horizontais para visualizar as despesas médias por cargo
fig, ax = plt.subplots(figsize=(16, 6))

# Desenha as barras horizontais
ax.barh(df_final['Cargo'], df_final['despesa_media'], color="#fa0707")
ax.invert_yaxis()  # Inverte a ordem das barras
ax.set_facecolor("#000000")  # Define a cor de fundo

# Título do gráfico
fig.suptitle("Despesas medias Por Cargo 2023")

# Adiciona uma grade com linhas brancas pontilhadas
plt.grid(color="white", linestyle=":", linewidth=0.5)

# Ajusta o tamanho das legendas no eixo Y
plt.yticks(fontsize=8)


# Adiciona uma fonte de referência no gráfico
plt.figtext(0.79, 0.89,"Fonte: Portal da transparencia", fontsize=7)

# Rótulo do eixo X
plt.xlabel("Despesas")

# Exibe o gráfico
plt.show()
```
